```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
premier_player_df = pd.read_csv("premier-player-23-24.csv")
key_passes_df = pd.read_csv("player_big_chances_created.csv")
team_goals_df = pd.read_csv("team_goals_per_match.csv")
team_possession_df = pd.read_csv("possession_percentage_team.csv")
team_shots_df = pd.read_csv("ontarget_scoring_att_team.csv")
```

```python
for df in [premier_player_df, key_passes_df, team_goals_df, team_possession_df, team_shots_df]:
    print(df.shape)
    print(df.columns)
    print(df.isnull().sum())
    print("-"*40)
```

```python
key_passes_df.rename(columns={
    'Big Chances Created':'big_chances_created',
    'Total Assists':'assists_total',
    'Minutes':'minutes_played',
    'Matches':'matches_played'}, inplace=True)

team_goals_df.rename(columns={
    'Goals per Match':'team_goals_per_match',
    'Total Goals Scored':'team_total_goals'}, inplace=True)

team_possession_df.rename(columns={'Possession (%)':'team_possession'}, inplace=True)

team_shots_df.rename(columns={
    'Shots on Target per Match':'team_shots_on_target',
    'Shot Conversion Rate (%)':'team_shot_conversion'}, inplace=True)
```

```python
player_data = premier_player_df.merge(key_passes_df[['Player','Team','big_chances_created','minutes_played','matches_played']], 
                                      on=['Player','Team'], how='inner')
```

```python
team_stats = team_goals_df.merge(team_possession_df[['Team','team_possession']], on='Team', how='inner')
team_stats = team_stats.merge(team_shots_df[['Team','team_shots_on_target','team_shot_conversion']], on='Team', how='inner')
```

```python
final_df = player_data.merge(team_stats, on='Team', how='inner')
```
